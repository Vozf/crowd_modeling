```python
import pickle
import holoviews as hv
from sklearn import datasets

hv.extension('matplotlib')
%matplotlib inline
```

```python
with open('google_news_embedding_1500.pkl','rb') as f:
    emb_news = pickle.load(f)
with open('s_curve_embedding_10000.pkl','rb') as f:
    emb_s = pickle.load(f)
with open('daily_mail_embeddings.pkl','rb') as f:
    emb_daily_mail = pickle.load(f)
```

```python
X, _ = datasets.make_s_curve(10000, random_state=0)
plot = hv.Scatter3D(X).opts(color='z', linewidth=0.00001)
hv.renderer('matplotlib').show(plot)
```

```python
for name, embedding in emb_news.items():
    plot = hv.Scatter((embedding[:, 0], embedding[:, 1])).opts(title=name)
    hv.renderer('matplotlib').show(plot)
```

```python
for name, embedding in emb_s.items():
    print(embedding.shape)
    plot = hv.Scatter((embedding[:, 0], embedding[:, 1])).opts(title=name)
    hv.renderer('matplotlib').show(plot)
```

```python
for name, embedding in emb_daily_mail.items():
    print(embedding.shape)
    plot = hv.Scatter((embedding[:, 0], embedding[:, 1])).opts(title=name)
    hv.renderer('matplotlib').show(plot)
```
